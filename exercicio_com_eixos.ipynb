# Trabalhando com eixos

### Importações

```python
import matplotlib.pyplot as plt
from cycler import cycler
```

### Gráfico de linhas

```python

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_xlabel("Mês")
ax.set_ylabel("Vendas (em milhares de reais)")
ax.set_title("Evolução de vendas ao longo do ano")

plt.show()
```

### Gráfico de barras

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

fig, ax = plt.subplots()

ax.bar(produtos, quantidades)

ax.set_xlabel("Produtos")
ax.set_ylabel("Quantidades de venda (em milhares de unidades)")
ax.set_title("Quantidades de vendas dos produtos da empresa F")

plt.show()
```

### Gráfico de pizza

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

fig, ax = plt.subplots()

ax.pie(quantidades, labels=categorias, explode=destaque, autopct='%.1f')


plt.show()
```

### Gráfico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label='Online')
ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
ax.legend()

ax.set_xlabel('Ano')
ax.set_ylabel('Vendas em milhões de reais')
ax.set_title('Comparação de vendas online x lojas')

ax.set_xticks(anos)

plt.show()
```

### Sistema de eixos em uma figura 
Obs: Com mosaico

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Gráfico 1 - Linha ############################################

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

axs[0, 0].plot(meses, vendas)
axs[0, 0].set_xlabel("Meses")
axs[0, 0].set_ylabel("Vendas (em milhares de reais)")
axs[0, 0].set_title("Evolução de vendas ao longo do ano")

# Gráfico 2 - Barras ##########################################

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

axs[0, 1].bar(produtos, quantidades)
axs[0, 1].set_xlabel("Produtos")
axs[0, 1].set_ylabel("Quantidades de venda (em milhares de unidades)")
axs[0, 1].set_title("Quantidades de vendas dos produtos da empresa F")


# Gráfico 3 - Pizza ############################################

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

axs[1, 0].pie(quantidades, labels=categorias, explode=destaque, autopct='%.1f')

# Gráfico 4 - Barras empilhadas #################################

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

axs[1, 1].bar(anos, vendas_online, label='Online')
axs[1, 1].bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
axs[1, 1].legend()

axs[1, 1].set_xlabel('Ano')
axs[1, 1].set_ylabel('Vendas em milhões de reais')
axs[1, 1].set_title('Comparação de vendas online x lojas')

axs[1, 1].set_xticks(anos)

##################################################################
fig.suptitle('Evolução de vendas', fontsize=16)

plt.show()
```

### Gráfico com mosaico

```python
mosaico = 'AA;BB;CD'
fig = plt.figure(figsize=(12, 14))
espacamento = {'wspace': 0.3, 'hspace':0.5}
axs = fig.subplot_mosaic(mosaico, gridspec_kw=espacamento)

# Gráfico 1 - Linha ############################################

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

axs['A'].plot(meses, vendas)
axs['A'].set_xlabel("Meses")
axs['A'].set_ylabel("Vendas (em milhares de reais)")
axs['A'].set_ylim(min(vendas) -5, max(vendas) + 5)
axs['A'].set_title("Evolução de vendas ao longo do ano")

# Gráfico 2 - Barras ##########################################

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

axs['B'].bar(produtos, quantidades)
axs['B'].set_xlabel("Produtos")
axs['B'].set_ylabel("Quantidades de venda (em milhares de unidades)")
axs['B'].set_title("Quantidades de vendas dos produtos da empresa")


# Gráfico 3 - Pizza ############################################

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

axs['C'].pie(quantidades, labels=categorias, explode=destaque, autopct='%.1f')

# Gráfico 4 - Barras empilhadas #################################

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

axs['D'].bar(anos, vendas_online, label='Online')
axs['D'].bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
axs['D'].legend()

axs['D'].set_xlabel('Ano')
axs['D'].set_ylabel('Vendas em milhões de reais')
axs['D'].set_title('Comparação de vendas online x lojas')

axs['D'].set_xticks(anos)

##################################################################
fig.suptitle('Evolução de vendas', fontsize=16)

plt.show()
```
