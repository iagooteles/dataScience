# Trabalhando com eixos



## Mais de um eixo y em uma figura

Crie uma figura de um gráfico de precipitação com um eixo secundário para a temperatura, para mostrar a relação entre a temperatura e a precipitação em uma cidade.



```python
import matplotlib.pyplot as plt
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]

cor_temperatura = 'tab:red'
cor_preciptacao = 'tab:blue'

fig, ax1 = plt.subplots(figsize=(16,8))

ax1.bar(meses, precipitacao_media, color=cor_preciptacao, label='Precipitação média em (mm)')
ax1.set_ylabel('Preciptação média (mm)', color=cor_preciptacao)
ax1.tick_params(axis='y', labelcolor=cor_preciptacao)
ax1.set_ylim(0, 225)

ax2 = ax1.twinx()

ax2.plot(meses, temperatura_media, color=cor_temperatura, linewidth=2, label='Precipitação média em (ºC)')
ax2.set_ylabel('Temperatura média (ºC)', color=cor_temperatura)
ax2.tick_params(axis='y', labelcolor=cor_temperatura)
ax2.set_ylim(20, 31)

ax1.set_title('Temperatura e precipitação ao longo do ano')
fig.legend(bbox_to_anchor=(0.8, 1), bbox_transform=ax1.transAxes)

plt.show()
```
