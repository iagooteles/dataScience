# Exercícios de mosaico de gráficos

Considere os dados da primeira célula para um aplicativo de uma pequena empresa de software. Faça um mosaico com 4 gráficos:

1. Gráfico de linhas com os novos usuários por mês
2. Gráfico de barras com a receita mensal
3. Gráfico de linhas com os usuários ativos diários
4. Gráfico de barras empilhadas com os tickets de suporte abertos e fechados

Os gráficos 1 e 2 devem ter uma linha própria no mosaico. Os gráficos 3 e 4 devem compartilhar a mesma linha.


```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_usuarios = [100, 80, 110, 140, 150, 160, 170, 120, 110, 150, 200, 140]
usuarios_ativos_diarios = [1000, 750, 900, 1300, 1400, 1500, 1200, 1200, 1500, 1900, 2000, 1800]
receita_mensal = [10000, 8500, 9800, 10000, 11000, 12000, 10000, 10500, 13000, 13000, 20000, 17000]
tickets_suporte_abertos = [110, 75, 100, 140, 170, 160, 100, 130, 190, 200, 210, 140]
tickets_suporte_fechados = [100, 85, 100, 120, 130, 150, 140, 110, 130, 120, 150, 200]
```

```python
mosaico = 'AA;BB;CD'

fig = plt.figure(figsize=(10, 14))
espacamento = {'wspace': 0.3, 'hspace':0.3}

axs = fig.subplot_mosaic(mosaico, gridspec_kw=espacamento)

# Gráfico de linhas #
axs['A'].plot(meses, novos_usuarios, marker='o')
axs['A'].set_ylabel('Novos usuários')
axs['A'].set_title('Novos usuários por mês')

# for i, valor in enumerate(novos_usuarios):
#     axs['A'].annotate(f'{valor}', (meses[i], novos_usuarios[i]), textcoords="offset points", xytext=(-15, -4), ha='center', fontsize=14)

# Gráfico de barras #
bars = axs['B'].bar(meses, receita_mensal)
axs['B'].bar_label(bars, fontweight='bold', color='white', padding=-20)
axs['B'].set_title('Receita mensal')

# Gráfico de linhas com usuários ativos diários #

axs['C'].plot(meses, usuarios_ativos_diarios, marker='o')
axs['C'].spines['top'].set_visible(False)
axs['C'].set_title('Usuários ativos diários')

for i, valor in enumerate(usuarios_ativos_diarios):
    axs['C'].annotate(f'{valor}', (meses[i], usuarios_ativos_diarios[i]), textcoords="offset points", xytext=(0, 15), ha='center', fontsize=8)

# Gráfico de barras empilhadas com os tickets de suporte abertos e fechados #

bars_tickets_abertos = axs['D'].bar(meses, tickets_suporte_abertos, label='Tickes suporte abertos')
bars_tickets_fechados = axs['D'].bar(meses, tickets_suporte_fechados, bottom=tickets_suporte_abertos, label='Tickets suporte fechados')
axs['D'].legend()
axs['D'].set_title('Tickets abertos x fechados')

axs['D'].bar_label(bars_tickets_abertos, color='white', padding=-20, fontsize=8)
axs['D'].bar_label(bars_tickets_fechados, color='white', padding=-20, fontsize=8)

############################
plt.suptitle('Empresa X')

plt.show()
```
