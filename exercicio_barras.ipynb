## Exercício - Gráfico de barras

Você deve criar um gráfico de barras com as calorias por 100 g de alimento. Os dados estão na primeira célula do Jupyter Notebook. O gráfico deve ter as seguintes características:

- Título: Calorias por 100 g de alimento => OK
- Título: O título do gráfico deve estar na cor cinza, em negrito e com tamanho de fonte 14 => OK
- Eixos: Alimento (eixo X) e Calorias (eixo Y) => OK
- Os nomes compostos dos alimentos devem ser quebrados em duas linhas => OK
- Cores: Todas as barras devem ser cinza => OK
- Transparência: Todas as barras devem ter transparência de 50% =>OK
- Cor de uma barra específica: A barra do bacon deve ser vermelha e ter transparência de 80% => OK
- Rótulos: Os rótulos de cada barra devem estar em negrito e na cor branca => OK
- Rótulos: Os rótulos de cada barra devem estar dentro da barra => OK
- Ticks: Os ticks do eixo Y devem ser removidos => OK
- Ticks: Os ticks do eixo X devem ser removidos => OK
- Box: A caixa do gráfico deve ser removida => OK
- Grid: O grid do gráfico deve ser removido => OK
- Figura: A figura deve ter tamanho 8x5 => OK
- Figura: A figura deve ser salva com o nome grafico_barras.png, com 300 dpi de resolução e com a área de plotagem ajustada => 
- Figura: A figura deve ser salva numa pasta chamada `imagens` que deve estar no mesmo diretório do Jupyter Notebook. => 

```python
import pandas as pd
import matplotlib.pyplot as plt

# valores por 100 g do alimento
calorias_por_alimento = {
    "Batata Frita": 607,
    "Batata Chips": 542,
    "Bacon": 533,
    "Pizza": 296,
    "Cachorro Quente": 260
}
```

```python
calorias_por_alimento_formatado = [categoria.replace(' ', '\n') for categoria in calorias_por_alimento.keys()]

plt.figure(figsize=(8,5))

bars = plt.bar(calorias_por_alimento_formatado, calorias_por_alimento.values(), color='grey', alpha=0.5)

bars[2].set_color("red")
bars[2].set_alpha(0.8)

plt.box(False)
plt.yticks([])

plt.tick_params(axis='x', length=0)

plt.bar_label(bars, fontweight='bold', color='white', padding=-20)

plt.title('Calorias por 100 g de alimento', fontdict={'fontsize': 14, 'fontweight': 'bold', 'color': 'gray'})

plt.savefig('images/grafico_barras.png', bbox_inches='tight', dpi=300)

plt.show()
```
